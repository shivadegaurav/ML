```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('train.csv')
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size=0.2,random_state=42)
```

```python
X_train.head()
```

```python
y_train.sample(5)
```

```python
trf1 = ColumnTransformer([('impute_age',SimpleImputer(),[2]),('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])],remainder='passthrough')
```

```python
trf2 = ColumnTransformer([('ohe_sex_embarked',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1,6])],remainder='passthrough')
```

```python
trf3 = ColumnTransformer([('scale',MinMaxScaler(),slice(0,10))])
```

```python
trf4 = SelectKBest(score_func=chi2,k=8)
```

```python
trf5 = DecisionTreeClassifier()
```

```python
#create pipeline 
pipe = Pipeline([('trf1',trf1),('trf2',trf2),('trf3',trf3),('trf4',trf4),('trf5',trf5)])
```

```python
#pipeline vs make_pipeline 
#pipeline requires naming of steps, make_pipelinne does not
#(Same applies to ColumnTransformer vs make_column_transformer)
```

```python
pipe = make_pipeline(trf1,trf2,trf3,trf4,trf5)
```

```python
pipe.fit(X_train,y_train)
```

```python
#explore the pipeline
pipe.named_steps
```

```python
from sklearn import set_config
set_config(display='diagram')
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#cross validation using pipeline
from sklearn.model_selection import cross_val_score
cross_val_score(pipe,X_train,y_train,cv=5,scoring='accuracy').mean()
```

```python
#grid search cv using pipeline
params={ 'decisiontreeclassifier__max_depth':[1,2,3,4,5,None]}
```

```python
from sklearn.model_selection import GridSearchCV
grid =GridSearchCV(pipe, params,cv=5,scoring='accuracy')
grid.fit(X_train,y_train)
```
